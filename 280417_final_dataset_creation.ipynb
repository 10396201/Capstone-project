```python
# replace negative numbers with 0
ita_pred[ita_pred < 0] = 0
ita_pred = ita_pred.round(decimals = 0)

est_pred[est_pred < 0] = 0
est_pred = est_pred.round(decimals = 0)
```

```python
ita = ita_pred
ita.Fri_1 = (ita.Fri_1 + ita.Fri_2)/2.0
ita.Sat_1 = (ita.Sat_1 + ita.Sat_2)/2.0
ita.Sun_1 = (ita.Sun_1 + ita.Sun_2)/2.0
ita.Mon_1 = (ita.Mon_1 + ita.Mon_2)/2.0
ita.Tue_1 = (ita.Tue_1 + ita.Tue_2)/2.0
ita.Wed_1 = (ita.Wed_1 + ita.Wed_2)/2.0
ita.Thu_1 = (ita.Thu_1 + ita.Thu_2)/2.0

ita = ita.ix[:,0:7]
ita = ita.round(decimals = 0)
```

```python
ita_days.head()

ita = ita.join(ita_days)
```

```python
def set_color(row, name):
    if row[name] < 4:
        return 'red'
    elif row[name] < 11:
        return 'yellow'
    else:
        return 'green'
```

```python
ita['col_mon'] = ita.apply (lambda row: set_color(row,'Mon_1'),axis=1)
ita['col_tue'] = ita.apply (lambda row: set_color(row,'Tue_1'),axis=1)
ita['col_wed'] = ita.apply (lambda row: set_color(row,'Wed_1'),axis=1)
ita['col_thu'] = ita.apply (lambda row: set_color(row,'Thu_1'),axis=1)
ita['col_fri'] = ita.apply (lambda row: set_color(row,'Fri_1'),axis=1)
ita['col_sat'] = ita.apply (lambda row: set_color(row,'Sat_1'),axis=1)
ita['col_sun'] = ita.apply (lambda row: set_color(row,'Sun_1'),axis=1)
```

```python
ita = ita[['Fri_1','Sat_1','Sun_1','Mon_1','Tue_1','Wed_1','Thu_1','col_mon','col_tue','col_wed','col_thu','col_fri','col_sat','col_sun']]
ita = ita.reset_index()

ita
```

```python
ita.to_csv('./Data/ita_plot.csv', index_label=False, index = False)
```
